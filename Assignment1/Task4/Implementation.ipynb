```python
import numpy as np
import random
```

```python
#transfrom adjacency to transition matrix (see lecture 10.3.1)
def adj_to_trans(matrix, tele_factor):
    print("\nconverting adjacency to transition matrix...")

    
    
    trans_matrix = [[] for i in range(len(matrix))]

    #keep index which rows were zero, so we can leave out later computation on them
    zero_index =[]

    #first divide elements by sum row
    row_nr=0
    for row in matrix:
        col_nr = 0
        row_sum = sum(row)
        for el in row:
            if row_sum==0:
                trans_matrix[row_nr].append(1 / len(row))
                zero_index.append(row_nr)
            else:
                trans_matrix[row_nr].append( matrix[row_nr][col_nr]/ row_sum)
            col_nr+=1
        row_nr+=1

    print("\ndivided each element by row sum, 0 rows set to 1/row_length")
    print(trans_matrix)

    #add teleportation factor to the matrix
    row_nr=0
    for row in matrix:
        if row_nr not in zero_index:
            col_nr = 0
            for col in matrix:
                trans_matrix[row_nr][col_nr] *= 1-tele_factor

                #could do other operation here as well, but we want to see every intermediate step
                #trans_matrix[row_nr][col_nr] += tele_factor/len(row)

                col_nr+=1
        row_nr+=1

    print("\nmultiplied matrix by 1-teleport_factor (except for original zero rows)")
    print(trans_matrix)

    # add teleportation factor to the matrix
    row_nr = 0
    for row in matrix:
        if row_nr not in zero_index:
            col_nr = 0
            for col in matrix:
                trans_matrix[row_nr][col_nr] += tele_factor / len(row)
                col_nr += 1
        row_nr += 1



    print("\nadded teleport_factor/row_length to every entry (except for original zero rows)")
    print(trans_matrix)
    return trans_matrix
```

```python
#compute probability vector based on adjacency matrix (see lecture 10.3.5)
def page_rank(adj_matrix, teleport_factor=0.1, probability_vector=None, iterations=20):

    trans_matrix = np.array(adj_to_trans(adj_matrix, teleport_factor))

    #if no vector was passed, create a random one with an arbitrarily placed 1
    if probability_vector==None:
        vec_len = len(adj_matrix)
        rnd = random.randint(0, vec_len-1)
        probability_vector = np.zeros(vec_len)
        probability_vector[rnd] = 1
    else:
        probability_vector = np.array(probability_vector)
        
    print("\n\ncomputing probability vector...")
    print("\nstarting vec: ", probability_vector)

    #multiply the vector with the transition matrix n times
    count=0
    while count<iterations:
        probability_vector = np.matmul(probability_vector, trans_matrix)
        print(count+1, "iteration:", probability_vector)
        count+=1

    return probability_vector
```
